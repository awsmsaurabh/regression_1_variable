```python
#importing required liberary
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#import datasets
train_data=pd.read_csv("train.csv");
test_data=pd.read_csv("test.csv");
```

```python
#removing Nan if any
train_data=train_data.dropna()
test_data=test_data.dropna()
```

```python
#loading training data in numpy array
x=np.array(train_data.loc[:,'x'])
y=np.array(train_data.loc[:,'y'])
```

```python
#loading testing data in numpy array
x_test=np.array(test_data.loc[:,'x'])
y_test=np.array(test_data.loc[:,'y'])
```

```python
#linear regression model
xy=x*y
xy_mean=xy.mean()
```

```python
x_mean=x.mean()
y_mean=y.mean()
```

```python
xx=x*x
xx_mean=xx.mean()
```

```python
#Calculating coefficient of eq y=ax+b
a=(xy_mean - x_mean*y_mean)/(xx_mean-x_mean*x_mean)
b=y_mean-a*x_mean
```

```python
print("a=",a)
print("b=",b)
```

```python
#predicting y
y_pred=a*x_test+b
```

```python
mean_sq_error=((y_pred-y_test)**2).sum()/len(y_test)
print("mean square error",mean_sq_error)
```

```python
#importing liberary for accuracy check
from sklearn.metrics import mean_squared_error, r2_score
print("Mean squaed error ",mean_squared_error(y_test,y_pred))
print("r2 score ",r2_score(y_test,y_pred))
```

```python
#Plotting training data
plt.scatter(x,y,color='red')
plt.title("Train data set")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#plotting test data
plt.scatter(x_test,y_test,color='red')
plt.plot(x_test,y_pred,color='blue')
plt.title("Test data set")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
